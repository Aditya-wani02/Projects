## import the data using api and convert it into a csv file

```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=d7a0f78cf708fe1d2c4d5adefa8c928d&language=en-US&page=1')
```

```python
pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=d7a0f78cf708fe1d2c4d5adefa8c928d&language=en-US&page={}'.format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df=df.append(temp_df,ignore_index=True)
```

## saving as a csv file

```python
df.to_csv('movies_data.csv')
```
